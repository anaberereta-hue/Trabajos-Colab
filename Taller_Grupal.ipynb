## **Parte 4: Análisis Exploratorio**


### 4.1 Estadística Descriptiva


```python
# 4.1.1 Promedio de ventas globales
media_global_sales= df['Global_Sales'].mean()
print(f"La media de ventas globales es: {media_global_sales}")
```

```python
# 4.1.2 Mediana
mediana_ventas_globales=df['Global_Sales'].median()
print(f"La mediana de las ventas globales es: {mediana_ventas_globales}")
```

```python
# 4.1.3 Genero con mayores ventas
ventas_genero = df.groupby('Genre')['Global_Sales'].sum()
print(f'asventas por genero son: {ventas_genero}')
MayorVentas_Genero = ventas_genero.idxmax()


print(f"El género con mayores ventas globales es: {MayorVentas_Genero}")
```

```python
# 4.1.4 Plataforma que aparece más frecuentemente
moda_plataforma=df['Platform'].mode()[0]
print(f"La plataforma que aparece mas frecuentemente es: {moda_plataforma}")
```

```python
# 4.1.5 Región con mayores ventas promedio
regiones = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
promedios = df[regiones].mean()
max_region = promedios.idxmax()
max_valor = promedios.max()

print(f"Promedios por región:\n{promedios}\n")
print(f"La región con mayores ventas promedio es {max_region} con {max_valor:.4f}")
```
